# Generators in python

    - generators are more memory efficient , compute efficient but slower iterators widely used to process large amout of data in sequantial manner.
    - Generators are also foundations to various frameworks in python for data loading , data processing and computing.
    - A generator can be created using yield keyword in a function in place of return keyword.
    - A generator can also created using tuple comprehension.

```python
x=[4,5,3,5,2]
print(x)
for i in x:
    print(i)
```

```python
for i in x:
    print(i)
```

```python
def myfun(a):
    return a+5,a+6,a+7,a+8,a+9
```

```python
out=myfun(3)
print(out)
```

```python
for i in out:print(i)
```
