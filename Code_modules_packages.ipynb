# Use of r before the string
- without  r we used \new then it takes as new line
- with r it prints same line

```python
x="Hello \new World"
print(x)
```

```python
z=r"Hello \new World" #Raw string 
print(z)              # that's why we used os.listdir(r"H:\Python_Programming") 'r' here
```
