#### Non-Primitive Data Types

### List
- Used to store collection of values, it supports repeatation of values
- A List supports values of multiple data types
- List is mutable - can be modified
- A list can be created in python by specifying a set of values inside square/upper/box bracket=[]

```python
x=[45,6,12.556,'Hello',True,12,True,45]
print(x)
type(x)
```

```python
# To check total number of elements in the list
len(x)
```

```python
# Indexing - Left to Right - Always starts with 0
print(x[0])  # First element
print(x[3])  # Fourth element
print(x[5])  # Sixth element
print(x[11])  # IndexError: list index out of range
```

```python
# Python supports negative indexing as well - right to left- starts with -1
print(x[-1])  # Last element
print(x[-4])  # Fourth last element
print(x[-8])  # Eighth last element
print(x[-9])  # IndexError: list index out of range
```

```python
# Slicing- Accessing a range of elements from the list
print(x[2:5]) # From index 2 to index 4 (5 is excluded) 
```

```python
print(x[0:4]) # From index 0 to index 3 (4 is excluded)
```

```python
print(x[-6:-1]) # From index -6 to index -2 (-1 is excluded)
# [45, 6, 12.556, 'Hello', True, 12, True, 45]
```

```python
# Modify a list element- List is mutable i.e. its elements can be changed
print(x)
x[3]='Python Programming'
print(x)
```

### Range() Vs SLice() Difference

```python
#Range() is used to generate a sequence of numbers(mostly in loops) 
#range(start,stop,step)

for i in range(1,10,2): # Ya Madhe 1 ha start aahe , 10 ha stop aahe ani 2 ha step aahe
                        # 2 step mhnje 1 ntr 2 step mhnje 3 ntr 2 step mhnje 5 ntr asa chalel
                        # if raange(1,10,3) mhnje 1 ntr 4 ntr 7 ntr asa chalel
                        #coz step  aahe 1 ntr 3 rd step mhnje 4 amd 4 ntr  rd step mhnje 7.
    print(i)
```

```python
#Slice() is used to extract a portion of a sequence like list, tuple, string etc.
#slice(start,stop,step)

data=[10,20,30,40,50,60]
s=slice(1,5,2)  # start=1, stop=5, step=2
                # index ithe 0 pasn start hoil mhnje index 1 la 20 aahes ani end la 5 mhnje index 4 prynt chalel
                # mhnje 20 pasn 50 prynt i.e. index 1 pasn index 4 prynt list bhetel
                # tya mdhe 2 chya step ne gheil mhnje output mdhe 20 ani 40 yeil 
print(data[s])  # Output: [20, 40]

# --- IGNORE ---# is slice(0,5,2) # start=0, stop=5, step=2
                 # index ithe 0 pasn start hoil mhnje index 0 la 10 aahes ani end la 5 mhnje index 4 prynt chalel
                # mhnje 10 pasn 50 prynt i.e. index 0 pasn index 4 prynt list bhetel
                # tya mdhe 2 chya step ne gheil mhnje output mdhe 10,30,50 yeil
```

#### Tuple
- Can be used to store a collection of elements
- Elements can be of different data types, it supports repeatation of values
- Tuples are immutable(cannot be modified or change)
- Tuples are created using paranthesis/small/lower/round bracket= ()

```python
y=(45,56,2.5,'Hello','hi',False)
print(y)
type(y)
```

```python
# Total number of elements in the tuple
len(y)
```

```python
#Indexing in tuple left to right - index starts from 0 same as list
print(y[0])  # First element
print(y[3])  # Fourth element
```

```python
# Indexing right to left - negative indexing from start -1
print(y[-1])  # Last element
print(y[-4])  # Fourth last element
```

```python
#SLicing in tuple
print(y[1:4])  # From index 1 to index 3 (4 is excluded)
```

```python
print(y[-5:-2])  # From index -5 to index -3 (-2 is excluded)
```

```python
# as Tuple is immutable i.e. its elements cannot be changed, we cannot modify tuple elements like lists.
y[2]=100  # This will raise a TypeError
```

#### Dictionary in python
- A collection of key value pairs stored as key:value
- Keys can be unique and immutable data type but values can be anything
- A dictionary is mutable object i.e. it can be modified
- Can be created using middle/curly/flower bracket= {} 

```python
mydict={"name":"Yogesh","age":25,"salary":5000}
print(mydict)
type(mydict)
```

```python
# Total number of elements in the dictionary using len()
len(mydict)
```

```python
#Indexing in dictionary- Accessing values using keys
print(mydict["name"])  # Access value associated with key 'name'
print(mydict["age"])   # Access value associated with key 'age'
print(mydict["salary"])   # Access value associated with key 'salary'
```

```python
#modify the dictionary values - modify values using keys
# replace/modify values for existing pair

mydict["name"]="Pranshu"
print(mydict)
```

```python
# Also we can add new key-value pairs to the dictionary
mydict["laptop"]="Dell"
print(mydict)
```

```python
# Remove a key-value pair using pop() method
mydict.pop("Salary")
print(mydict)
```

```python
# print keys and values separately
print(mydict.keys())  # Print all keys
print(mydict.values())  # Print all values
```

```python
# List inside the dictionary example
m={"name":["John","Jacob","James","Jenny","Jessy"],"age":[25,28,37,38,34],"salary":[5000,6000,5500,7000,8000]}
print(m)
type(m)
```

```python
# It will gives me all the names present in the list associated with the key "name"
m["name"]
```

```python
# for speficis name in the list inside the dictionary
m["name"][2] # Access the third name in the list associated with the key "name"
```

```python
# add pair again in it
m["department"]=["HR","Finance","IT","Admin","Sales"]
print(m)
```

```python
# Also add one more name in the list associated with the key "name" using append() method
m["name"].append("Jasmine")
print(m)
```

```python
# Also insert one more name at specific position in the list associated with the key "name" using insert() method
m["name"].insert(2,"Jazzy")
print(m)
```

#### Set in Python
- A collection of values, but it does not support repeatation of values
- Supports values of different data types
-  Sets are mutable i.e. can be modified
- it is unordered- we can not do indexing
- Widely used for set theory operations - union, intersection
- can be created using curly/middle/flower brackets = {}

```python
k={45,12,'hi',5.2,10}
print(k)
type(k)
```

```python
len(k)
```

```python
a={4,5,2,3,6,5,8}
b={8,9,3,1,2,7,8,9}
print(a)
print(b)
```

```python
# Union of two sets means all unique elements from both sets
a.union(b)
```

```python
# Intersection of two sets means common elements in both sets
a.intersection(b)
```

```python
# To add value to a set use add() method
a.add(10)
print(a)
```

```python
# To remove value from a set use remove() method
a.remove(4)
print(a)
```

```python
# To remove any random value from the set use pop() method
a.pop()
print(a)
```

```python
a={4,5,6,7,2,3,8}
b={1,2,3,7,8,9}
print(a)
print(b)
```

```python
a.intersection_update(b) # It will keep only the common elements in set 'a' that are also present in set 'b' rest elemets are removed from set 'a'

```

```python
print(a)
print(b)
```

```python
a.difference(b) # It will remove all the elements from set 'a' that are also present in set 'b' and keep the rest of the elements in set 'a' intact
print(a)
```

```python
a.difference_update(b) # It will remove all the elements from set 'a' that are also present in set 'b' and keep the rest of the elements in set 'a' intact.
print(a)
print(b)
```

```python
a={1,2}
b={3,4}
c={1,2,5,6,4}
a.union(b).intersection(c)
```
