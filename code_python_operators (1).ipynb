### Operators in Python

#### Arithmetic Operators

```python
x=5
y=6
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x//y) # // is used for floor division in python
print(x%y)  # % is used for modulus in python
print(x**y) # ** is used for power in python
```

#### Logical Operators

```python
a=True
b=False
print(a and b)
print(a or b)
print(not a)
print(not b)
```

#### Membership Operators

- in
- not in

```python
'Hi' in "Hi from Yogesh" # returns True cause Hi is present in the string
```

```python
"Hello" in "Hi from Yogesh"  # returns False cause Hello is not present in the string
```

```python
k=[8,5,4,8,6,3]
4 in k  # returns True cause 4 is present in the list
```

```python
12 in k  # returns False cause 12 is not present in the list
```
