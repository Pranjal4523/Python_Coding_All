#### Range

#### Reverse built in function


```python
x=[4,2,5,8,7,2]
reversed(x) #Again it gives a generator object as a output
```

```python
for k in reversed(x):
  print(k)
```

```python
list(reversed(x)) # first we need to convert it into list and then we can see the output
```

#### Zip

```python
x=[4,5,2,3,8]
y=[7,2,5,8,9]
zip(x,y) # Again it gives generator object as a output
```

```python
list(zip(x,y)) #After converting to list then it gives zipping elements as a ouput
```

```python
x.extend(y)
```

```python
x # Extendd the list means added the list y to x
```

#### Map Function in ![Python](path)
  - Used for Data Transformation


```python
x=[4,5,2,3,6,8]

def myfun(a):
    return a+5
```

```python
myfun(4) # It takes it as 4+5 and print 9 only
```

```python
# A Normal Approach
z=[myfun(k) for k in x]
print(z)
```

```python
## Using Map function first it created generator object
y=map(myfun,x)
```

```python
type(y) #Again it gives map i.e. return as generator type of object
```

```python
for k in y:   # For print just used for loop on y generator object
    print(k)   
```

```python
list(map(myfun,x)) ## Or convert it into list and then print
```
