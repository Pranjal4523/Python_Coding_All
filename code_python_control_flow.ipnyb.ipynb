### Task- BMI Calculator

    - ask user to enter the weight in kg
    - ask user to enter the height in cm

    -Convert height from cm to meteres

    Use below formula to calculate BMI
    BMI= w/(h*h)
    print(BMI)

```python
weight= int(input("Enter your weight in kgs: "))
height= int(input("Enter your height in ceentimeteres: "))

height=height/100  # converting cm to meters

bmi= weight/(height*height)

print("Your BMI is:",round(bmi,2))
```

#### Loops in Python
- Loops are used to repeat a set of codes mulitple times

    - for loop - whenever exact count of repeatation is known
    - while loop - Whenever the exact count of repeatation is not known, rather we want to perform condition based repeatation

```python
print("Python is easy and amazing")
print("Python is easy and amazing")
print("Python is easy and amazing")
print("Python is easy and amazing")
print("Python is easy and amazing")
```

```python
for i in range(5):
    print("Python is easy and amazing")
```

```python
for i in range(5):
    print("Python is easy and amazing",i)
```

```python
for i in range(3,8):
    print("Python is easy and amazing",i)
```

```python
for i in range(3,8,2):
    print("Python is easy and amazing",i)
```

```python
# for reverse order
for i in range(20,2,-2):
    print("Python is awesome",i)
```

```python
# For loop also use iterate through the list

mylist=["DEL","DXB","MUM","CHE","BAN","BOM","AUH","AMD","MAA","DDN"]
mylist.sort(reverse=True)
for city in mylist:
    print("My city is =",city)
```
