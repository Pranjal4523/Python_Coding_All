### User Defined Functions

#### Format 

    - def function_name(input arg1,input arg2,...);
        <function defination - operation>
        return <out_arg> , <out_arg2>

```python
def myfun():
    print("This is my function")
    
```

```python
myfun() #calling the function
```

```python
## With Parameters
def myfun2(a,b):
    c=a+b
    return c
```

```python
myfun2(a=5,b=10) #calling function with parameters
```

```python
myfun2(10,20) #calling function with positional arguments
```

```python
# We can call fucntion in same cell in jupyter notebook or in different cells
def myfun2(a,b):
    c=a+b
    return c
myfun2(b=10,a=25) #calling function with keyword arguments
```

```python
myfun2(8) #calling function with missing argument (will give error)
```

### Default value for input argument - Optional Argument

```python
def myfun3(a,b=5):
    c=a+b
    return c
myfun3(10) #calling function with one argument (b will take default value)
```

```python
myfun3(6,7) #calling function with both arguments (b will take given value)
```

```python
m=myfun3(6,2)
print(m) # Storing return value in variable and printing it
x=5     # As requested, no additional text is provided beyond the code snippet and return statement only return values after that any code then it not works then if need to work properly then store it in third variable and return that variable
```

```python
# Function can also return multiple values
def myfun4(a,b):
    c=a+b
    d=a*b
    return c,d
```

```python
myfun4(10,30) #calling function which returns multiple values
```

```python
x,y=myfun4(a=4,b=6) #storing multiple return values in different variables
print(x)
print(y)
```

```python
z=myfun4(3,7) #storing multiple return values in a single variable
print(z) # it will be a tuple containing both return values
```

#### A function with varialble number of arguments

```python
def m1(a,b):
    c=a+b
    return c
```

```python
m1(7,3)
```

```python
m1(7,3,2)
```

```python
def m1(a,*b):
    print(a,type(a))
    print(b,type(b))
```

```python
m1(5) #calling function with single argument
```

```python
m1(5,2) #calling function with multiple arguments
```

```python
m1(5,2,4,7,8,9) #calling function with multiple arguments i.e.vaiable length arguments
```

```python
def m1(a,*b):
    c=a+sum(b)
    return c
```

```python
m1(5,2,4,7,8,9) 
```

#### Variable number of keyword arguments

```python
def fun1(a,**b): #function with variable length keyword arguments
    print(a,type(a)) #a is an integer
    print(b,type(b)) #b is a dictionary
    return None
```

```python
fun1(12)
```

```python
fun1(12,name="Ashish") #calling function with variable length keyword arguments
```

```python
fun1(12,name="John",age=30,city="New York",salary=50000)
```
